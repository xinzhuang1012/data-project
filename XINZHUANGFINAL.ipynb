```python
def summarize_performance(threshold):
    df_test['FPRED']=(df_test['FPROB']>=threshold)*1
    CT=pd.crosstab(df_test['FPRED'],df_test['FOREC'])
    print(CT)
    print("\n")
    cpos=CT.iloc[0,1]+CT.iloc[1,1]
    cneg=CT.iloc[0,0]+CT.iloc[1,0]
    ppos=CT.iloc[1,0]+CT.iloc[1,1]
    pneg=CT.iloc[0,0]+CT.iloc[0,1]
    TPR=CT.iloc[1,1]/cpos # recall, sensitivity
    FPR=CT.iloc[1,0]/cneg # false positive rate
    PPV=CT.iloc[1,1]/ppos # precision/positive pred value
    FOR=CT.iloc[0,1]/pneg # false ommision rate
    print("true positive rate = %5.3f" % (TPR))
    print("false positive rate = %5.3f" % (FPR))
    print("positive predictive value = %5.3f" % (PPV))
    print("false omission rate = %5.3f" % (FOR))
```
